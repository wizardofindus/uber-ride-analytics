```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
pd.set_option("display.max_columns", None)
```

```python
# Reading data from CSV dataset:

df = pd.read_csv(r"C:\Users\USER\Documents\uber_trips_dataset_50k.csv")
```

```python
#Data Cleaning:

#Convert columns to Datetime:

df['pickup_time'] = pd.to_datetime(df['pickup_time'])

df['drop_time'] = pd.to_datetime(df['drop_time'])
```

```python
df.info()
```

```python
#Finding null(missing) values:

df.isnull().sum()
```

```python
# removing null records if any:

df.dropna()
```

```python
# Removing Invalid Coordinates:

df = df[
    (df['pickup_lng'].between(-180, 180)) &
    (df['pickup_lat'].between(-90, 90))
]
```

```python
# Removing Outliers (Fare):

df = df[(df['fare_amount'] > 0) & (df['fare_amount'] < 200)]
```

```python
# Feature Engineering:

df['hour'] = df['pickup_time'].dt.hour
df['day'] = df['pickup_time'].dt.day
df['month'] = df['pickup_time'].dt.month
df['day_of_week'] = df['pickup_time'].dt.day_name()
```

```python
# Time Category:

def time_of_day(hour):
    if 5 <= hour < 12:
        return "Morning"
    elif 12 <= hour < 17:
        return "Afternoon"
    elif 17 <= hour < 21:
        return "Evening"
    else:
        return "Night"

df['time_category'] = df['hour'].apply(time_of_day)
```

```python
# Distance Calculation (Haversine Formula):

from math import radians, cos, sin, sqrt, atan2

def haversine(lat1, lon1, lat2, lon2):
    R = 6371
    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)
    
    a = sin(dlat/2)**2 + cos(radians(lat1)) * cos(radians(lat2)) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1-a))
    
    return R * c

df['distance_km'] = df.apply(lambda row: haversine(
    row['pickup_lat'], row['pickup_lng'],
    row['dropoff_lat'], row['dropoff_lng']
), axis=1)
```

```python
# Fare per KM:

df['fare_per_km'] = df['fare_amount'] / df['distance_km']

```

```python
# Peak Hour Flag:

df['peak_hour'] = df['hour'].apply(lambda x: 1 if x in [8,9,17,18,19] else 0)
```

```python
# Advanced Feature Engineering:

# Trip Category:

df['trip_type'] = pd.cut(
    df['distance_km'],
    bins=[0,2,5,15,100],
    labels=['Short','Medium','Long','Very Long']
)
```

```python
# High Fare Flag:

df['high_fare'] = np.where(df['fare_amount'] > df['fare_amount'].quantile(0.75), 1, 0)
```

```python
# Demand Index:

hourly_demand = df['hour'].value_counts(normalize=True)

df['demand_index'] = df['hour'].map(hourly_demand)
```

```python
# Export cleaned dataset:

df.to_csv("uber_cleaned.csv", index=False)
```
